```python
import pandas as pd

input_file = 'TCE-PB-Receitas-2018_2024.csv'

df = pd.read_csv(input_file, sep=';', encoding='latin-1')

df.head(3)
```

```python
df_2024 = df[df['data_arrecadacao'].apply(lambda x: True if '2024' in x else False)]
```

```python
df_2024.reset_index(inplace=True)
```

```python
df_2024.to_csv('receita_2024.csv', sep=';', index=False)
```
